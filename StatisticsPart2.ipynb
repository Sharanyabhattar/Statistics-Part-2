1. What is hypothesis testing in statistics?
- Hypothesis testing is a formal way in statistics to make decisions or inferences about a population based on sample data.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
-  Null Hypothesis (H‚ÇÄ):
It is the default assumption or status quo.

Assumes no effect, no difference, or no relationship exists.

It's what you assume to be true unless proven otherwise.

Example phrases:

"There is no difference between groups."

"The average is equal to X."

"There is no relationship between variables."

- Where as Alternative Hypothesis (H‚ÇÅ or Ha):
This is the opposite of the null hypothesis.

It claims that there is an effect, difference, or relationship.

It‚Äôs what you aim to support through your test.

Example phrases:

"There is a difference between groups."

"The average is not equal to X."

"There is a relationship between variables."

3. What is the significance level in hypothesis testing, and why is it important?
- .The significance level, denoted as Œ± (alpha), is the threshold probability used to decide whether to reject the null hypothesis.

   .It represents the maximum acceptable risk of making a Type I error (rejecting a true null hypothesis).

- Why is it Important?
 - Controls Error Risk:

It limits the chance of false positives (Type I error).

For example, Œ± = 0.05 means you're willing to accept a 5% chance of wrongly rejecting H‚ÇÄ when it's actually true.

 - Decision Boundary:

It sets the cutoff for p-values.

If p ‚â§ Œ±, reject H‚ÇÄ (result is statistically significant).

If p > Œ±, fail to reject H‚ÇÄ (result is not statistically significant).

 - Helps Interpret Results:

It gives context to your p-value.

A result with a p-value of 0.03 is considered significant if Œ± = 0.05, but not significant if Œ± = 0.01.

4. What does a P-value represent in hypothesis testing?
- The p-value is the probability of observing your sample data (or more extreme data), assuming the null hypothesis (H‚ÇÄ) is true.

- It measures how compatible your data is with H‚ÇÄ.

- A small p-value suggests that the observed data is unlikely under H‚ÇÄ, giving you reason to doubt H‚ÇÄ.

5. How do you interpret the P-value in hypothesis testing?
- Simple Interpretation:
The p-value tells you how likely your data (or something more extreme) is, assuming the null hypothesis (H‚ÇÄ) is true.

Smaller p-value ‚Üí Stronger evidence against H‚ÇÄ.

- Step-by-Step Interpretation:
Set a Significance Level (Œ±):

Example: Œ± = 0.05 (common choice).

Compare p-value to Œ±:

If p-value ‚â§ Œ±:

Reject H‚ÇÄ.

The result is statistically significant.

There is evidence to support the alternative hypothesis (H‚ÇÅ).

If p-value > Œ±:

Fail to reject H‚ÇÄ.

The result is not statistically significant.

There's not enough evidence to support H‚ÇÅ, but that doesn‚Äôt prove H‚ÇÄ is true.

6.  What are Type 1 and Type 2 errors in hypothesis testing?
- Type 1 Error (False Positive):
What is it?

Rejecting the null hypothesis (H‚ÇÄ) when it is actually true.

In simple words:

You think there is an effect or difference, but actually, there isn't.

Risk controlled by:

Significance level (Œ±).

If Œ± = 0.05 ‚Üí 5% chance of Type I error.

Example:
You conclude a drug works when it actually doesn't.

- Type 2 Error (False Negative):
What is it?

Failing to reject the null hypothesis (H‚ÇÄ) when it is actually false.

In simple words:

You miss a real effect or difference.

Risk represented by:

Œ≤ (beta).

The power of the test = 1 - Œ≤ (probability of detecting a true effect).

Example:
You conclude a drug doesn't work when it actually does.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
- One-tailed Test (Directional Test)
Tests for an effect in one specific direction only.

You are only interested if the parameter is greater than OR less than a certain value, not both.

Alternative hypothesis (H‚ÇÅ):

Right-tailed: H‚ÇÅ: Œº > Œº‚ÇÄ

Left-tailed: H‚ÇÅ: Œº < Œº‚ÇÄ

Example:
A company claims a new battery lasts more than 10 hours.

H‚ÇÄ: Œº ‚â§ 10 hours

H‚ÇÅ: Œº > 10 hours (right-tailed test)

When to use:
When theory or logic tells you only one direction is meaningful or relevant.

- Two-tailed Test (Non-directional Test)
Tests for an effect in both directions.

You check if the parameter is either greater than OR less than a certain value.

Alternative hypothesis (H‚ÇÅ):

H‚ÇÅ: Œº ‚â† Œº‚ÇÄ

Example:
You want to check if a machine produces parts with a diameter different from 5 cm (could be bigger or smaller).

H‚ÇÄ: Œº = 5 cm

H‚ÇÅ: Œº ‚â† 5 cm (two-tailed test)

When to use:
When you're looking for any difference, regardless of direction.

3. Data is approximately normally distributed

. important for small samples.

10.  What is the T-distribution, and when should it be used instead of the normal distribution?
- What is the T-distribution?
The T-distribution (Student's T-distribution) is a bell-shaped, symmetric distribution, similar to the normal distribution.

It is used when the sample size is small and/or the population standard deviation (œÉ) is unknown.

The T-distribution has thicker tails than the normal distribution.

This accounts for extra uncertainty from estimating the population SD using the sample SD (s).
 | Situation |                        Use T-distribution? | Use Z-distribution |

|----------|                          ----------|             ----------|

| Population Standard deviation (œÉ)       | Yes   |                 No   |

| Small sample size (n < 30)              | Yes   |                 No   |

| Large sample size (n ‚â• 30) & œÉ known    | Yes  |                  No   |

| Population SD known (rare in real life) | No   |                  Yes   |

11. What is the difference between a Z-test and a T-test?
- Z-test assumes perfect knowledge of the population SD and is ideal when sample sizes are big enough that variability is minimized.

- T-test is designed for small samples and when you don't know the population SD, which is more realistic.

- As sample size increases, the T-distribution becomes similar to the normal distribution, meaning both tests will give nearly the same result.

12.  What is the T-test, and how is it used in hypothesis testing?
- The T-test is a statistical test used to determine whether there is a significant difference between means.

It uses the T-distribution, which accounts for extra uncertainty due to small sample sizes and unknown population standard deviation (œÉ).

- How is it used in hypothesis testing?
1. Set up hypotheses:

- H‚ÇÄ (Null hypothesis): No difference (e.g., Œº = 50).

- H‚ÇÅ (Alternative hypothesis): There is a difference (e.g., Œº ‚â† 50).

2. Calculate the T-statistic using the formula.

3. Determine the degrees of freedom (df):

- For one-sample t-test: df = n - 1.

4. Find the critical t-value from t-distribution tables (depends on df and chosen Œ± level, e.g., 0.05).

5. Decision:

- If |t-statistic| > critical t ‚Üí Reject H‚ÇÄ.

- Else ‚Üí Fail to reject H‚ÇÄ.

13. What is the relationship between Z-test and T-test in hypothesis testing?
- Both are used for comparing means
Both Z-test and T-test are inferential statistical tests used to test hypotheses about population means.

They help you determine whether an observed sample statistic is significantly different from a known population parameter or from another sample.

- They converge as sample size grows
As sample size increases (n ‚â• 30 or larger):

The T-distribution approaches the standard normal distribution.

The difference between the T-test and Z-test becomes negligible.

In practice, for large samples, either test gives almost the same result.

- Simple analogy:
Think of Z-test as the ideal case with complete information, while the T-test is more realistic when dealing with incomplete or estimated information (small samples, unknown œÉ).

14.  What is a confidence interval, and how is it used to interpret statistical results?
- A confidence interval (CI) is a range of values that is likely to contain the true population parameter (like the mean or proportion) with a certain level of confidence (e.g., 95%, 99%).

- Interpretation of CI:
We are 95% confident that the true population mean lies within this interval.

It does NOT mean there is a 95% probability that the mean is in the interval (since the population mean is fixed).

It means if we repeated the sampling process many times, about 95% of the calculated intervals would contain the true mean.

15. What is the margin of error, and how does it affect the confidence interval?
- The Margin of Error (MOE) is the "plus or minus" part of a confidence interval (CI).
It quantifies the amount of uncertainty in the estimate due to sampling variability.
- The Margin of Error (MOE) directly controls the width of the confidence interval (CI).
- The wider the MOE, the wider the confidence interval, meaning less precision.

- The smaller the MOE, the narrower the confidence interval, meaning more precision.

16. How is Bayes' Theorem used in statistics, and what is its significance?
- Bayes' Theorem is a way to update our beliefs (probabilities) about something after seeing new evidence. Instead of only relying on initial assumptions, Bayes' theorem lets us revise those assumptions based on actual data.

- Bayes' theorem is powerful when you have some prior knowledge and you collect new data.

. It lets you update your estimates dynamically as new data comes in.

. It‚Äôs widely used in fields like medicine, finance, and machine learning.

17.  What is the Chi-square distribution, and when is it used?
- The Chi-square (œá¬≤) distribution is a probability distribution that is widely used in hypothesis testing and inferential statistics, especially when dealing with categorical data.

It is positively skewed (starts at 0 and goes to infinity) and its shape depends on the degrees of freedom (df).
As df increases, the distribution becomes more symmetric and bell-shaped.

- When is the Chi-square distribution used?
1. Testing for Goodness of Fit

- Purpose: To test if an observed frequency distribution matches an expected distribution.

- Example: Does a die show all numbers equally often?

2. Testing for Independence (Chi-square test of independence)

- Purpose: To test if two categorical variables are independent or associated.

- Example: Is there a relationship between gender and product preference?

3. Testing Homogeneity

- Purpose: To test if different populations have the same distribution across categories.

- Example: Do voters from different regions show the same political party preference?

4. Invariance of Variance Tests (less common use)

- Chi-square is also used to test variance or standard deviation against a hypothesized value when data is normal.

18. What is the Chi-square goodness of fit test, and how is it applied?
- The Chi-square Goodness of Fit test is a statistical test used to determine if a sample data set matches an expected distribution.
- it is applied When you have one categorical variable with multiple categories,
also to check if the observed data fit a theoretical or expected distribution (e.g., uniform, normal, historical proportions).

19. What is the F-distribution, and when is it used in hypothesis testing?
- The F-distribution is a continuous probability distribution that arises when comparing two sample variances. It is positively skewed (right-skewed) and depends on two sets of degrees of freedom:

Numerator degrees of freedom (df‚ÇÅ)

Denominator degrees of freedom (df‚ÇÇ)

The F-distribution is used to model the ratio of two scaled chi-square distributions, which makes it ideal for comparing variances.

- When is the F-distribution used in hypothesis testing?
1. Comparing Variances (F-test)

- To test if two populations have equal variances.

- Example: Comparing the variability of test scores between two different classrooms.

2. Analysis of Variance (ANOVA)

- To test if means of three or more groups are equal.

- ANOVA uses the F-distribution to compare between-group variance to within-group variance.

- Example: Testing if different teaching methods produce different average test scores.

3. Regression Analysis

- To test the overall significance of a regression model.

- It compares the model fit with no predictors to the model with predictors.

20. What is an ANOVA test, and what are its assumptions?
- ANOVA (Analysis of Variance) is a statistical method used to compare the means of three or more groups to see if at least one group mean is significantly different from the others.

Instead of doing multiple t-tests (which increases error), ANOVA tests all groups simultaneously.

It assesses whether variability between group means is larger than variability within groups.

- Assumptions of ANOVA:
1. Independence of observations
Data points in different groups are independent of each other.

2. Normality
The residuals (differences between observations and group means) are approximately normally distributed within each group.

3. Homogeneity of variances (Homoscedasticity)
The variances across the groups are roughly equal.

21. What are the different types of ANOVA tests?
-  Different types of ANOVA tests
1. One-Way ANOVA
Purpose: Compare means across one independent categorical variable with three or more groups.

Example: Comparing test scores of students from three different teaching methods.

2. Two-Way ANOVA (Factorial ANOVA)
Purpose: Compare means across two independent categorical variables simultaneously, and also test for interaction effects between them.

Example: Comparing test scores based on teaching method (3 types) and gender (2 types), and seeing if gender changes the effect of teaching.

3. Repeated Measures ANOVA
Purpose: Compare means when the same subjects are measured multiple times (within-subjects design).

Example: Measuring blood pressure of patients before, during, and after treatment.

4. Mixed-Design ANOVA
Purpose: Combines between-subjects factors (different groups) and within-subjects factors (repeated measures).

Example: Testing the effect of two diets (between-subjects) on weight loss measured monthly (within-subjects).

5. MANOVA (Multivariate ANOVA)
Purpose: Like ANOVA but compares multiple dependent variables simultaneously.

Example: Comparing the effect of a teaching method on both math and reading scores together.

22.  What is the F-test, and how does it relate to hypothesis testing?
- The F-test is a statistical test that uses the F-distribution to compare variances or to test the overall significance of a model.

It essentially tests whether the variability between groups is significantly greater than the variability within groups.

- How does the F-test work in hypothesis testing?
  - Null Hypothesis (H‚ÇÄ): The group variances (or means, depending on the test) are equal‚Äîno effect or no difference.

  - Alternative Hypothesis (H‚ÇÅ): At least one group variance (or mean) differs.

```python
# 2.  Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range for the discrete uniform distribution
low = 1    # inclusive
high = 7   # exclusive (this gives values 1 through 6)

# Create the distribution object
dist = randint(low, high)

# Values (support) over which to evaluate the PMF
x = np.arange(low, high)
pmf = dist.pmf(x)

# Plotting the PMF
# The 'use_line_collection' argument is deprecated and removed in newer matplotlib versions.
plt.stem(x, pmf, basefmt=" ")
plt.title("PMF of Discrete Uniform Distribution (1 to 6)")
plt.xlabel("Values")
plt.ylabel("Probability")
plt.ylim(0, 1)
plt.grid(True)
plt.show()
```

```python
# 3.  Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
def bernoulli_pdf(x, p):
    """
    Compute the PDF (PMF) of a Bernoulli distribution.

    Parameters:
    - x: 0 or 1 (the outcome)
    - p: probability of success (0 <= p <= 1)

    Returns:
    - Probability of observing x
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1.")

    return p if x == 1 else (1 - p)

# Example usage
print("P(X=1) with p=0.7:", bernoulli_pdf(1, 0.7))
print("P(X=0) with p=0.7:", bernoulli_pdf(0, 0.7))

```

```python
# 6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.
import numpy as np
import matplotlib.pyplot as plt

def discrete_uniform_cdf(a, b):
    # Support values
    x = np.arange(a, b + 1)
    n = b - a + 1  # Number of discrete points

    # Probability mass function (PMF) - equal probability for each outcome
    pmf = np.ones(n) / n

    # CDF is the cumulative sum of the PMF
    cdf = np.cumsum(pmf)

    return x, cdf

# Example parameters
a = 1
b = 10

x, cdf = discrete_uniform_cdf(a, b)

# Plotting the CDF
plt.step(x, cdf, where='post', label='CDF')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('CDF of Discrete Uniform Distribution')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 7. Generate a continuous uniform distribution using NumPy and visualize it.
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the continuous uniform distribution
a = 0  # lower bound
b = 5  # upper bound
size = 1000  # number of samples

# Generate random samples
samples = np.random.uniform(a, b, size)

# Define x values for the theoretical PDF
x = np.linspace(a - 1, b + 1, 500)
pdf = np.where((x >= a) & (x <= b), 1 / (b - a), 0)  # PDF of uniform distribution

# Plot histogram of samples with density=True to normalize
plt.hist(samples, bins=30, density=True, alpha=0.6, color='skyblue', label='Sampled data')

# Plot theoretical PDF
plt.plot(x, pdf, 'r-', lw=2, label='Theoretical PDF')

plt.title('Continuous Uniform Distribution')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#9. Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import matplotlib.pyplot as plt

def plot_z_scores(data):
    # Calculate mean and standard deviation
    mean = np.mean(data)
    std = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.figure(figsize=(8, 4))
    plt.plot(z_scores, marker='o', linestyle='-', color='teal')
    plt.axhline(0, color='gray', linestyle='--', linewidth=1)
    plt.title('Z-scores of Dataset')
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.grid(True)
    plt.show()

    return z_scores

# Example usage
data = np.array([10, 12, 23, 23, 16, 23, 21, 16])
z_scores = plot_z_scores(data)
```

```python
#10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
population_size = 100000
sample_sizes = [1, 5, 30, 100]  # Different sample sizes to show CLT effect
num_samples = 1000  # Number of samples to draw for each sample size

# Generate population data from an exponential distribution (non-normal)
population = np.random.exponential(scale=1.0, size=population_size)

plt.figure(figsize=(12, 8))

for i, n in enumerate(sample_sizes, 1):
    # Draw num_samples samples of size n and compute their means
    sample_means = [np.mean(np.random.choice(population, n, replace=False)) for _ in range(num_samples)]

    # Plot histogram of sample means
    plt.subplot(2, 2, i)
    plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='black')

    # Overlay the normal distribution curve with the same mean and std dev
    mean = np.mean(population)
    std = np.std(population) / np.sqrt(n)  # std dev of the sample mean

    x = np.linspace(min(sample_means), max(sample_means), 100)
    plt.plot(x, norm.pdf(x, mean, std), 'r-', lw=2, label='Normal PDF')

    plt.title(f'Sample Size = {n}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
#14.  Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    """Calculate the Z-score."""
    return (x - mean) / std

def plot_z_score(x, mean, std):
    """Calculate Z-score and visualize it on the standard normal curve."""
    z = calculate_z_score(x, mean, std)

    # Plot settings
    x_vals = np.linspace(-4, 4, 1000)
    y_vals = norm.pdf(x_vals, 0, 1)

    plt.figure(figsize=(10, 5))
    plt.plot(x_vals, y_vals, label='Standard Normal Distribution', color='blue')

    # Shade area under curve up to Z
    x_shade = np.linspace(-4, z, 1000)
    y_shade = norm.pdf(x_shade, 0, 1)
    plt.fill_between(x_shade, y_shade, color='skyblue', alpha=0.5, label=f'P(Z ‚â§ {z:.2f})')

    # Draw vertical line at Z
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')

    plt.title('Z-score on Standard Normal Distribution')
    plt.xlabel('Z')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Output
    print(f"Z-score for x = {x} (mean = {mean}, std = {std}) is: {z:.2f}")
    print(f"Cumulative probability (P(Z ‚â§ {z:.2f})) = {norm.cdf(z):.4f}")

# Example usage
x = 75
mean = 70
std = 10

plot_z_score(x, mean, std)
```

```python
#15.  Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
from scipy.stats import norm

def z_test(sample, mu_0, sigma, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)

    # Calculate Z statistic
    z = (sample_mean - mu_0) / (sigma / np.sqrt(n))

    # Two-tailed test: critical z value
    z_critical = norm.ppf(1 - alpha/2)

    # p-value
    p_value = 2 * (1 - norm.cdf(abs(z)))

    # Decision
    reject_null = abs(z) > z_critical

    # Output
    print(f"Sample mean = {sample_mean:.4f}")
    print(f"Z-statistic = {z:.4f}")
    print(f"Z-critical (¬±) = {z_critical:.4f}")
    print(f"P-value = {p_value:.4f}")
    print("Result: ", "Reject H‚ÇÄ" if reject_null else "Fail to reject H‚ÇÄ")

    return z, p_value, reject_null

# Example usage:
# Known population std dev
sigma = 15

# Hypothesized population mean
mu_0 = 100

# Sample data (can be randomly generated or real)
sample_data = np.array([102, 98, 105, 100, 95, 108, 99, 101, 103, 97])

z_stat, p_val, decision = z_test(sample_data, mu_0, sigma)
```

```python
#16. Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_critical * std_err
    lower = mean - margin_of_error
    upper = mean + margin_of_error

    print(f"Sample mean = {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval = ({lower:.2f}, {upper:.2f})")
    print("Interpretation: We are", int(confidence*100),
          "% confident that the true population mean lies within this interval.")

    return (lower, upper)

# Example usage
sample_data = np.array([102, 98, 105, 100, 95, 108, 99, 101, 103, 97])
ci = confidence_interval(sample_data, confidence=0.95)
```

```python
#3.  Implement a one-sample Z-test using Python to compare the sample mean with the population mean.
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    standard_error = pop_std / np.sqrt(n)

    # Calculate Z-score
    z_score = (sample_mean - pop_mean) / standard_error

    # Calculate two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Print results
    print(f"Sample Mean = {sample_mean:.4f}")
    print(f"Population Mean = {pop_mean}")
    print(f"Z-score = {z_score:.4f}")
    print(f"P-value = {p_value:.4f}")

    # Hypothesis test conclusion
    if p_value < alpha:
        print(f"Reject the null hypothesis at significance level Œ± = {alpha}")
        print("There is sufficient evidence to conclude the sample mean differs from the population mean.")
    else:
        print(f"Fail to reject the null hypothesis at significance level Œ± = {alpha}")
        print("There is insufficient evidence to conclude the sample mean differs from the population mean.")

# Example usage:
np.random.seed(1)
pop_mean = 100
pop_std = 15  # known population standard deviation

# Generate sample data with mean slightly different from population mean
sample_data = np.random.normal(loc=102, scale=pop_std, size=40)

one_sample_z_test(sample_data, pop_mean, pop_std)
```

```python
#4.Perform a two-tailed Z-test using Python and visualize the decision region on a plot
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    std_err = pop_std / np.sqrt(n)

    # Calculate Z statistic
    z_stat = (sample_mean - pop_mean) / std_err

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    print(f"Sample Mean = {sample_mean:.4f}")
    print(f"Population Mean = {pop_mean}")
    print(f"Z-statistic = {z_stat:.4f}")
    print(f"P-value = {p_value:.4f}")

    # Critical Z values for two-tailed test
    z_critical = norm.ppf(1 - alpha/2)

    # Plotting
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Standard Normal Distribution')

    # Shade rejection regions
    plt.fill_between(x, 0, y, where=(x <= -z_critical), color='red', alpha=0.4, label='Rejection Region (left)')
    plt.fill_between(x, 0, y, where=(x >= z_critical), color='red', alpha=0.4, label='Rejection Region (right)')

    # Mark the Z statistic
    plt.axvline(z_stat, color='blue', linestyle='--', label=f'Z-statistic = {z_stat:.2f}')

    # Annotate
    plt.title('Two-tailed Z-test Decision Regions')
    plt.xlabel('Z value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Decision interpretation
    if abs(z_stat) > z_critical:
        print("Reject the null hypothesis: sample mean significantly differs from population mean.")
    else:
        print("Fail to reject the null hypothesis: insufficient evidence to say sample mean differs.")

# Example usage
np.random.seed(123)
pop_mean = 100
pop_std = 15

sample = np.random.normal(loc=105, scale=pop_std, size=30)

two_tailed_z_test(sample, pop_mean, pop_std)
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(pop_mean, pop_std, alt_mean, sample_size, alpha=0.05):
    # Standard error
    se = pop_std / np.sqrt(sample_size)

    # Critical Z-value for two-tailed test
    z_critical = norm.ppf(1 - alpha/2)

    # Critical values on sample mean scale
    crit_low = pop_mean - z_critical * se
    crit_high = pop_mean + z_critical * se

    # Define range for plotting
    x = np.linspace(pop_mean - 4*se, alt_mean + 4*se, 1000)

    # Null distribution pdf
    null_pdf = norm.pdf(x, loc=pop_mean, scale=se)
    # Alternative distribution pdf
    alt_pdf = norm.pdf(x, loc=alt_mean, scale=se)

    plt.figure(figsize=(12, 6))

    # Plot null distribution
    plt.plot(x, null_pdf, label='Null Hypothesis (H0)', color='blue')
    # Plot alternative distribution
    plt.plot(x, alt_pdf, label='Alternative Hypothesis (H1)', color='green')

    # Shade Type 1 error regions (alpha)
    x_alpha_left = np.linspace(pop_mean - 4*se, crit_low, 200)
    plt.fill_between(x_alpha_left, 0, norm.pdf(x_alpha_left, pop_mean, se), color='red', alpha=0.4, label='Type 1 Error (Œ±)')

    x_alpha_right = np.linspace(crit_high, pop_mean + 4*se, 200)
    plt.fill_between(x_alpha_right, 0, norm.pdf(x_alpha_right, pop_mean, se), color='red', alpha=0.4)

    # Shade Type 2 error region (beta)
    x_beta = np.linspace(crit_low, crit_high, 200)
    plt.fill_between(x_beta, 0, norm.pdf(x_beta, alt_mean, se), color='orange', alpha=0.5, label='Type 2 Error (Œ≤)')

    # Plot critical values
    plt.axvline(crit_low, color='black', linestyle='--', label='Critical Values')
    plt.axvline(crit_high, color='black', linestyle='--')

    plt.title('Type 1 and Type 2 Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Calculate Type 1 error (should be alpha)
    print(f"Type 1 error (Œ±): {alpha:.3f}")

    # Calculate Type 2 error (beta): probability alt distribution falls in acceptance region
    beta = norm.cdf(crit_high, loc=alt_mean, scale=se) - norm.cdf(crit_low, loc=alt_mean, scale=se)
    power = 1 - beta
    print(f"Type 2 error (Œ≤): {beta:.3f}")
    print(f"Statistical Power (1 - Œ≤): {power:.3f}")

# Example usage
visualize_type1_type2_errors(pop_mean=100, pop_std=15, alt_mean=110, sample_size=30, alpha=0.05)
```

```python
#6.  Write a Python program to perform an independent T-test and interpret the results.
import numpy as np
from scipy.stats import ttest_ind

def independent_t_test(group1, group2, alpha=0.05):
    # Perform independent two-sample t-test (Welch's t-test by default)
    t_stat, p_value = ttest_ind(group1, group2, equal_var=False)

    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpret the result
    if p_value < alpha:
        print(f"Reject the null hypothesis at Œ± = {alpha}.")
        print("There is a statistically significant difference between the two groups.")
    else:
        print(f"Fail to reject the null hypothesis at Œ± = {alpha}.")
        print("There is no statistically significant difference between the two groups.")

# Example usage
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=5, size=30)
group2 = np.random.normal(loc=53, scale=5, size=30)

independent_t_test(group1, group2)
```

```python
#7.Perform a paired sample T-test using Python and visualize the comparison results.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel, sem, t

# Perform paired t-test and visualize
def paired_t_test_visualize(before, after, alpha=0.05):
    # Calculate differences
    differences = after - before

    # Perform paired t-test
    t_stat, p_value = ttest_rel(after, before)

    # Calculate confidence interval for mean difference
    mean_diff = np.mean(differences)
    se_diff = sem(differences)
    df = len(differences) - 1
    t_crit = t.ppf(1 - alpha/2, df)
    ci_low = mean_diff - t_crit * se_diff
    ci_high = mean_diff + t_crit * se_diff

    # Print results
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Mean difference: {mean_diff:.4f}")
    print(f"{100*(1-alpha)}% confidence interval of difference: [{ci_low:.4f}, {ci_high:.4f}]")

    # Visualization
    plt.figure(figsize=(8,6))

    # Plot paired points with lines
    for i in range(len(before)):
        plt.plot([1, 2], [before[i], after[i]], marker='o', color='gray', alpha=0.5)

    # Plot means
    plt.scatter(1, np.mean(before), color='blue', label='Before', s=100)
    plt.scatter(2, np.mean(after), color='red', label='After', s=100)

    # Plot mean difference with confidence interval
    plt.errorbar(2.5, mean_diff, yerr=[[mean_diff - ci_low], [ci_high - mean_diff]],
                 fmt='o', color='green', capsize=5, label='Mean difference ¬± CI')

    plt.xticks([1, 2, 2.5], ['Before', 'After', 'Difference'])
    plt.title('Paired Sample T-Test Results')
    plt.ylabel('Values')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example data
np.random.seed(0)
before = np.random.normal(loc=50, scale=5, size=20)
after = before + np.random.normal(loc=2, scale=5, size=20)  # some improvement

paired_t_test_visualize(before, after)
```

```python
#8. Simulate data and perform both Z-test and T-test, then compare the results using Python
import numpy as np
from scipy.stats import norm, ttest_1samp

def one_sample_z_test(sample, pop_mean, pop_std):
    n = len(sample)
    sample_mean = np.mean(sample)
    std_err = pop_std / np.sqrt(n)
    z_stat = (sample_mean - pop_mean) / std_err
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    return z_stat, p_value

def one_sample_t_test(sample, pop_mean):
    t_stat, p_value = ttest_1samp(sample, pop_mean)
    return t_stat, p_value

# Simulate data
np.random.seed(2025)
pop_mean = 100
pop_std = 15  # Known population std deviation
sample_size = 30

# Generate sample with mean shifted from pop_mean
sample = np.random.normal(loc=103, scale=pop_std, size=sample_size)

# Perform Z-test
z_stat, z_p = one_sample_z_test(sample, pop_mean, pop_std)

# Perform T-test
t_stat, t_p = one_sample_t_test(sample, pop_mean)

print(f"One-sample Z-test results:")
print(f"Z-statistic = {z_stat:.4f}, p-value = {z_p:.4f}\n")

print(f"One-sample T-test results:")
print(f"T-statistic = {t_stat:.4f}, p-value = {t_p:.4f}\n")

# Interpretation
alpha = 0.05
print("Interpretation at alpha = 0.05:")
print(f"Z-test: {'Reject H0' if z_p < alpha else 'Fail to reject H0'}")
print(f"T-test: {'Reject H0' if t_p < alpha else 'Fail to reject H0'}")
```

```python
#9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
from scipy.stats import t

def confidence_interval(sample, confidence=0.95):
    n = len(sample)
    mean = np.mean(sample)
    sem = np.std(sample, ddof=1) / np.sqrt(n)  # standard error of mean
    df = n - 1  # degrees of freedom

    # t critical value for two-tailed test
    t_crit = t.ppf((1 + confidence) / 2, df)

    margin_of_error = t_crit * sem
    lower = mean - margin_of_error
    upper = mean + margin_of_error

    return (lower, upper)

# Example usage
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=30)

ci_lower, ci_upper = confidence_interval(data, confidence=0.95)
print(f"95% Confidence interval: [{ci_lower:.2f}, {ci_upper:.2f}]")
```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data.
import numpy as np
from scipy.stats import t

def margin_of_error(sample, confidence=0.95):
    n = len(sample)
    sem = np.std(sample, ddof=1) / np.sqrt(n)  # Standard error of the mean
    df = n - 1  # Degrees of freedom

    # t critical value for two-tailed test
    t_crit = t.ppf((1 + confidence) / 2, df)

    margin = t_crit * sem
    return margin

# Example usage
np.random.seed(123)
sample_data = np.random.normal(loc=50, scale=10, size=40)

moe = margin_of_error(sample_data, confidence=0.95)
print(f"Margin of Error at 95% confidence: {moe:.4f}")
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
def bayesian_inference(prior_disease, sensitivity, specificity):
    # Prior probabilities
    P_D = prior_disease
    P_not_D = 1 - P_D

    # Likelihoods
    P_Pos_given_D = sensitivity
    P_Pos_given_not_D = 1 - specificity

    # Total probability of positive test
    P_Pos = P_Pos_given_D * P_D + P_Pos_given_not_D * P_not_D

    # Posterior probability using Bayes' Theorem
    P_D_given_Pos = (P_Pos_given_D * P_D) / P_Pos

    return P_D_given_Pos

# Example values:
prior = 0.01        # 1% prevalence of disease
sensitivity = 0.95  # 95% chance test positive if disease present
specificity = 0.90  # 90% chance test negative if no disease

posterior_prob = bayesian_inference(prior, sensitivity, specificity)
print(f"Probability of having the disease given a positive test: {posterior_prob:.4f}")
```

```python
#12.Perform a Chi-square test for independence between two categorical variables in Python.
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Example data: contingency table of two categorical variables
# Suppose we survey 100 people about two variables: Gender (Male/Female) and Preference (Like/Dislike)
data = np.array([[30, 10],  # Males: 30 Like, 10 Dislike
                 [20, 40]]) # Females: 20 Like, 40 Dislike

# Create a DataFrame for better readability
df = pd.DataFrame(data, columns=['Like', 'Dislike'], index=['Male', 'Female'])
print("Contingency Table:")
print(df)

# Perform Chi-square test for independence
chi2, p, dof, expected = chi2_contingency(data)

print(f"\nChi-square statistic: {chi2:.4f}")
print(f"Degrees of freedom: {dof}")
print(f"P-value: {p:.4f}")

print("\nExpected frequencies if independent:")
print(pd.DataFrame(expected, columns=['Like', 'Dislike'], index=['Male', 'Female']))

# Interpretation
alpha = 0.05
if p < alpha:
    print("\nResult: Reject null hypothesis ‚Äî variables are associated (dependent).")
else:
    print("\nResult: Fail to reject null hypothesis ‚Äî variables are independent.")
```

```python
#13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.
import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test
    :param observed: 2D numpy array or similar with observed counts
    :return: 2D numpy array with expected frequencies
    """
    observed = np.array(observed)
    row_totals = observed.sum(axis=1, keepdims=True)
    col_totals = observed.sum(axis=0, keepdims=True)
    total = observed.sum()

    expected = row_totals @ col_totals / total  # Matrix multiplication

    return expected

# Example observed data (contingency table)
observed_data = np.array([[30, 10],
                          [20, 40]])

expected_freq = calculate_expected_frequencies(observed_data)

# Display as DataFrame for readability
df_expected = pd.DataFrame(expected_freq, columns=['Like', 'Dislike'], index=['Male', 'Female'])
print("Expected Frequencies:")
print(df_expected)
```

```python
#14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
import numpy as np
from scipy.stats import chisquare

# Observed frequencies (example)
observed = np.array([8, 10, 9, 12, 11, 10])

# Expected frequencies assuming fair die (uniform distribution)
total_rolls = observed.sum()
expected = np.full_like(observed, total_rolls / len(observed))

# Perform chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: observed data significantly differs from expected distribution.")
else:
    print("Fail to reject null hypothesis: observed data fits the expected distribution.")
```
